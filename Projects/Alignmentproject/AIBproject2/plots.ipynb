```python
import matplotlib.pyplot as plt
import numpy as np

def plot_data(Affine, Linear, step_length):
	x_values = list(range(100, 100 + step_length*len(Affine), step_length))
	x_values_np = np.array(x_values)
	y_values_np = x_values_np**2
	y_over_x_values_affine = Affine / x_values_np
	y_over_x_values_linear = Linear / x_values_np

	fig = plt.figure(figsize=(10, 10))

	ax1 = fig.add_subplot(221)
	ax1.plot(x_values, Affine, color='tab:red')
	ax1.set_title('Affine')
	ax1.set_xlabel('n')
	ax1.set_ylabel('Seconds')

	ax2 = fig.add_subplot(222)
	ax2.plot(x_values_np, y_over_x_values_affine, color='tab:blue', linestyle='--')
	ax2.set_title('time/n for Affine')
	ax2.set_xlabel('n')
	ax2.set_ylabel('Seconds/n')

	ax3 = fig.add_subplot(223)
	ax3.plot(x_values, Linear, color='tab:green')
	ax3.set_title('Linear')
	ax3.set_xlabel('n')
	ax3.set_ylabel('Seconds')

	ax4 = fig.add_subplot(224)
	ax4.plot(x_values_np, y_over_x_values_linear, color='tab:orange', linestyle='--')
	ax4.set_title('Seconds/n for Linear')
	ax4.set_xlabel('n')
	ax4.set_ylabel('Seconds/n')

	plt.tight_layout()
	plt.show()
```
